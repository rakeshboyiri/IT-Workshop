```python
label=["Date","Elderberry","Apple","Banana","Cherry"]
x=np.array([10,20,15,30,25])
c=["red","yellow","blue","orange","green"]
plt.subplot(2,1,1)
plt.title("Pie Chart of Fruit Distribution")

plt.pie(x,labels=label,colors=c,autopct='%1.0f%%')

x = np.arange(0,3*np.pi,0.1)
y = np.log(x+1)
e = np.exp(x/10)
plt.subplot(2,1,2)
plt.title("Line Plot of exp(x/10) and log(x+1)")
plt.xlabel("x")
plt.ylabel("Values")
plt.plot(x,y,color="green")
plt.plot(x,e)
plt.legend(["log(x+1)","exp(x/10)"])
plt.show()
```
