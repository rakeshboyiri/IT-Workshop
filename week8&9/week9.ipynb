```python
fruits = ["apple","banana","grapes","orange"]
count = [10,20,30,40]
plt.pie(count,explode=[0.1,0,0,0],shadow=False,labels=fruits,autopct="%.1f")
plt.legend(fruits)
plt.title("Sells analysis")
plt.show()
```
